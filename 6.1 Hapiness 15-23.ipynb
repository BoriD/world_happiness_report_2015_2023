### Creating a table containing all yearly happiness data

```python
#create a year column with the respective year in each table
y_15['Year']= 2015
y_16['Year']= 2016
y_17['Year']= 2017
y_18['Year']= 2018
y_19['Year']= 2019
y_20['Year']= 2020
y_21['Year']= 2021
y_22['Year']= 2022
y_23['Year']= 2023
```

```python
#Create a list of dataframes
tables = [y_15, y_16, y_17, y_18, y_19, y_20, y_21, y_22, y_23] 
```

```python
#Use pd.concat() to concatenate them vertically (union) 
happiness = pd.concat(tables, ignore_index=True)
```

### Consistency check


```python
#Looking for mixed type data
for col in happiness.columns.tolist():
  weird = (happiness[[col]].applymap(type) != happiness[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (happiness[weird]) > 0:
    print (col)
```

No mixed type data was found.
